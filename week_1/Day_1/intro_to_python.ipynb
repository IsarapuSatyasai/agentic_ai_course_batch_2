## 1. Variables and Data Types

**Explanation:**  
Variables are used to store data. Python is dynamically typed, meaning you don't need to declare the type.  
Common data types: `str` (text), `int` (whole numbers), `float` (decimal numbers), `bool` (True/False).

```python
# Creating variables
name = "Alice"        # String - used for text
age = 25              # Integer - whole numbers
height = 5.75         # Float - decimal numbers
is_student = True     # Boolean - True or False

# Printing variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)

# Checking data types
print("\n--- Checking Data Types ---")
print("Type of name:", type(name))
print("Type of age:", type(age))
print("Type of height:", type(height))
print("Type of is_student:", type(is_student))

# Type Conversion (Type Casting)
print("\n--- Type Conversion ---")
print("Age as string:", str(age))
print("Height as integer:", int(height))
```

## 2. Basic Operators

**Explanation:**  
Operators are used to perform operations on variables and values.  
- Arithmetic: `+`, `-`, `*`, `/`, `//`, `%`, `**`  
- Comparison: `==`, `!=`, `>`, `<`, `>=`, `<=`  
- Logical: `and`, `or`, `not`

```python
a = 15
b = 4

print("--- Arithmetic Operators ---")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division (no decimal):", a // b)
print("Modulus (remainder):", a % b)
print("Exponent (power):", a ** 2)

print("\n--- Comparison Operators ---")
print("Is a greater than b?", a > b)
print("Is a equal to b?", a == b)
print("Is a not equal to b?", a != b)

print("\n--- Logical Operators ---")
print("Both conditions true?", (a > 10) and (b < 5))
print("At least one true?", (a > 20) or (b < 5))
```

## 3. Control Flow (Decision Making)

**Explanation:**  
`if`, `elif`, and `else` allow your program to make decisions based on conditions.  
Python uses indentation (4 spaces) to define code blocks.

```python
score = 82

if score >= 90:
    grade = "A"
    print("Excellent!")
elif score >= 80:
    grade = "B"
    print("Very Good!")
elif score >= 70:
    grade = "C"
    print("Good!")
else:
    grade = "D"
    print("Needs Improvement!")

print(f"\nYour score is {score}. Your grade is {grade}.")
```

## 4. Loops

**Explanation:**  
Loops are used to repeat a block of code multiple times.

- `for` loop: Used when you know how many times to repeat.
- `while` loop: Used when you want to repeat until a condition becomes False.

```python
print("--- For Loop Example ---")
for i in range(1, 6):          # range(1,6) means 1 to 5
    print(f"Count: {i}")

print("\n--- While Loop Example ---")
count = 0
while count < 5:
    print("Learning Python is fun!")
    count += 1          # Important: increment to avoid infinite loop
```

## 5. Basic Functions

**Explanation:**  
Functions are reusable blocks of code.  
- `def` keyword is used to define a function.  
- You can pass parameters and return values.  
- Functions help organize and reuse code.

```python
# Simple function
def greet(name):
    """This function greets the person"""
    print(f"Hello, {name}! Welcome to Python class.")

# Function with return value
def add_numbers(x, y):
    return x + y

# Function with default parameter
def introduce(name, age=20):
    print(f"My name is {name} and I am {age} years old.")

# Calling the functions
greet("Bob")
greet("Sarah")

result = add_numbers(10, 15)
print("Sum is:", result)

introduce("Emma")
introduce("Liam", 28)   # overriding default age
```
