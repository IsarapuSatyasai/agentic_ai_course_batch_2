## 1. Lists

**Definition**: Lists are ordered, mutable sequences that can hold items of different data types.

**Key Characteristics**:
- Ordered (maintains insertion order)
- Mutable (can be modified after creation)
- Allows duplicate elements
- Dynamic size (can grow and shrink)
- Implemented as dynamic arrays internally

**Common Operations**:
- `append()`, `insert()`, `extend()`
- `remove()`, `pop()`, `clear()`
- `index()`, `count()`, `sort()`, `reverse()`
- Slicing and list comprehensions

**Advantages**:
- Very flexible and versatile
- Excellent for most general-purpose collections

**Disadvantages**:
- Slow for frequent insertions/deletions in the middle (O(n))
- Slower membership testing (`in` operator) compared to sets

**Use Cases**:
- Storing sequences of data
- Implementing stacks and queues
- Data processing and transformations

```python
# Detailed List Examples
fruits = ["apple", "banana", "cherry", "apple"]

# Adding elements
fruits.append("orange")
fruits.insert(2, "mango")
fruits.extend(["grape", "kiwi"])

print("Final list:", fruits)
print("Length:", len(fruits))
print("Count of 'apple':", fruits.count("apple"))

# Removing elements
fruits.remove("banana")      # removes first occurrence
popped = fruits.pop(3)       # removes and returns element at index
print("After removals:", fruits)

# Slicing and comprehension
print("Every other item:", fruits[::2])

numbers = [1, 5, 3, 9, 2]
squared = [x**2 for x in numbers if x > 3]
print("Squared (filtered):", squared)
```

## 2. Tuples

**Definition**: Tuples are ordered, immutable sequences.

**Key Characteristics**:
- Ordered and immutable (cannot change after creation)
- Allows duplicate elements
- More memory efficient than lists
- Hashable (can be used as dictionary keys)

**Common Operations**:
- Indexing and slicing (same as lists)
- Unpacking
- `count()` and `index()`

**Advantages**:
- Faster than lists
- Safer (data cannot be accidentally modified)
- Can be used as dictionary keys

**Disadvantages**:
- Cannot modify once created
- Limited methods compared to lists

**Use Cases**:
- Fixed configuration data
- Returning multiple values from functions
- Coordinates, RGB values, database records

```python
# Detailed Tuple Examples
coordinates = (10.5, 20.8, 15.2)
person = ("Alice", 25, "Engineer")

print("Tuple:", coordinates)
print("First element:", coordinates[0])

# Tuple unpacking
x, y, z = coordinates
print(f"X: {x}, Y: {y}, Z: {z}")

# Multiple return values
def circle_info(radius):
    return (3.14 * radius**2, 2 * 3.14 * radius)  # area, circumference

area, circumference = circle_info(5)
print(f"Area: {area:.2f}, Circumference: {circumference:.2f}")
```

## 3. Sets

**Definition**: Sets are unordered collections of unique elements.

**Key Characteristics**:
- Unordered
- Mutable
- No duplicate elements
- Highly optimized for membership testing

**Common Operations**:
- `add()`, `remove()`, `discard()`
- Mathematical operations: `union(|)`, `intersection(&)`, `difference(-)`, `symmetric_difference(^)`

**Advantages**:
- Extremely fast membership testing (`in` operator is O(1) average)
- Automatic duplicate removal
- Powerful set algebra

**Disadvantages**:
- No order (cannot do indexing or slicing)
- Elements must be hashable (immutable)

**Use Cases**:
- Removing duplicates from data
- Fast lookup / membership checks
- Finding common or unique items between collections

```python
# Detailed Set Examples
numbers = {1, 2, 3, 3, 4, 5, 5}
print("Set (duplicates removed):", numbers)

# Membership testing
print("Is 3 present?", 3 in numbers)

# Set operations
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

print("Union:", set_a | set_b)
print("Intersection:", set_a & set_b)
print("Difference (A - B):", set_a - set_b)
print("Symmetric Difference:", set_a ^ set_b)

# Adding / Removing
numbers.add(10)
numbers.discard(3)   # safer than remove()
print("Modified set:", numbers)
```

## 4. Dictionaries (dict)

**Definition**: Dictionaries are mutable mappings of unique keys to values.

**Key Characteristics**:
- Keys must be immutable (hashable)
- Values can be any type
- Fast average O(1) lookup time
- Maintains insertion order (Python 3.7+)

**Common Operations**:
- `get()`, `setdefault()`, `update()`
- `keys()`, `values()`, `items()`
- Dictionary comprehensions

**Advantages**:
- Extremely fast lookups
- Highly flexible for structured data

**Disadvantages**:
- Higher memory usage
- Keys must be unique and hashable

**Use Cases**:
- Storing configuration settings
- Mapping IDs to objects
- Counting with manual logic
- JSON-like data structures

```python
# Detailed Dictionary Examples
student = {
    "name": "Alice",
    "age": 21,
    "courses": ["Math", "Physics", "Chemistry"],
    "active": True
}

print("Student name:", student["name"])
print("Courses:", student["courses"])

# Safe access
print("Grade:", student.get("grade", "Not assigned"))

# Updating
student["age"] = 22
student.update({"city": "New York", "gpa": 3.8})

# Iteration
for key, value in student.items():
    print(f"{key}: {value}")

# Dictionary comprehension
squares = {x: x*x for x in range(1, 6)}
print("Squares:", squares)
```

## 5. NamedTuple

**Definition**: Factory function for creating tuple subclasses with named fields.

**Key Characteristics**:
- Immutable like regular tuples
- Fields are accessible by name (more readable)
- Still lightweight and memory efficient

**Advantages**:
- More readable than plain tuples
- Self-documenting
- Can be used wherever tuples are used

**Use Cases**:
- Data records with clear field names
- Replacing simple classes for better performance
- Structured function returns

```python
from collections import namedtuple

# Creating NamedTuple
Point = namedtuple('Point', ['x', 'y', 'z'])
Color = namedtuple('Color', 'r g b')

p1 = Point(10, 20, 30)
red = Color(255, 0, 0)

print("Point:", p1)
print("X coordinate:", p1.x)      # access by name
print("As tuple:", tuple(p1))     # can still use as tuple

# Function return
def get_employee():
    Employee = namedtuple('Employee', ['id', 'name', 'department', 'salary'])
    return Employee(1001, "Sarah", "Engineering", 95000)

emp = get_employee()
print(f"{emp.name} works in {emp.department} and earns ${emp.salary}")
```

## 6. defaultdict

**Definition**: A dictionary subclass that calls a factory function to supply missing values.

**Key Characteristics**:
- Same as dict but provides default values
- Avoids KeyError for missing keys

**Use Cases**:
- Grouping data into lists or other structures
- Counting with initialization
- Building nested dictionaries easily

```python
from collections import defaultdict

# Example 1: Grouping
words = ["apple", "banana", "apple", "cherry", "banana", "date"]

group = defaultdict(list)
for word in words:
    group[word].append(word)

print("Grouped words:", dict(group))

# Example 2: Default integer (for counting)
count = defaultdict(int)
for word in words:
    count[word] += 1

print("Word frequency:", dict(count))
```

## 7. Counter

**Definition**: A dict subclass specially designed for counting hashable objects.

**Key Characteristics**:
- Automatically counts occurrences
- Provides useful methods like `most_common()`

**Use Cases**:
- Word frequency analysis
- Finding most/least common items
- Statistical analysis of data

```python
from collections import Counter

text = """
Python is a popular programming language. 
Python is used for web development, data science, and automation.
"""

word_count = Counter(text.lower().split())
print("Top 5 most common words:", word_count.most_common(5))

# Arithmetic with Counters
c1 = Counter(a=3, b=1, c=2)
c2 = Counter(a=1, b=3, d=4)

print("Combined:", c1 + c2)
print("Subtracted:", c1 - c2)
```

## 8. deque (Double-ended Queue)

**Definition**: A generalization of stacks and queues optimized for fast appends and pops from both ends.

**Key Characteristics**:
- Fast O(1) operations at both ends
- Can have a maximum length (automatically discards old items)

**Use Cases**:
- Implementing queues (FIFO) and stacks (LIFO)
- Maintaining recent history or sliding windows
- When you need fast insertions/deletions from both ends

```python
from collections import deque

# Basic usage
d = deque(['task1', 'task2', 'task3'])
d.append('task4')        # add to right
d.appendleft('task0')    # add to left

print("Deque:", d)

print("Pop right:", d.pop())
print("Pop left:", d.popleft())

# Fixed size buffer (very useful!)
history = deque(maxlen=4)
for i in range(6):
    history.append(i)
    print(f"History after {i}: {list(history)}")
```

## When to Use Which? — Agentic AI Cheat Sheet

| Data Structure     | Best Use Case in Agents                        | Why It's Valuable                              |
|--------------------|------------------------------------------------|------------------------------------------------|
| **List**           | Full reasoning trace, action history           | Simple, ordered, easy to append                |
| **NamedTuple**     | Structured tool results & observations         | Readability + immutability                     |
| **Set**            | Visited entities, deduplication                | Fast membership testing                        |
| **Dict**           | Core agent state, long-term memory             | Flexible key-value storage                     |
| **defaultdict**    | Memory grouped by category/tool                | Automatic initialization                       |
| **Counter**        | Tool usage analytics, entity frequency         | Quick insights into agent behavior             |
| **deque**          | Rolling short-term memory, task queues         | Efficient fixed-size context window            |

**Pro Tip**: Many production agents combine these:
- `defaultdict(deque)` → Categorized rolling memory
- `dict + NamedTuple` → Structured world model
