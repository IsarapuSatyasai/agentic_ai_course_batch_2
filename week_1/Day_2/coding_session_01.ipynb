**Problem 1: Grade Calculator with Validation**

Write a program that asks the user to enter a score between 0 and 100. 
- If the input is invalid (not a number or out of range), show an error message.
- Otherwise, print the grade (A, B, C, D, F) and a short remark.

```python
try:
    score = float(input("Enter your score (0-100): "))
    
    if score < 0 or score > 100:
        print("Error: Score must be between 0 and 100.")
    elif score >= 90:
        print("Grade: A | Excellent Performance!")
    elif score >= 80:
        print("Grade: B | Very Good!")
    elif score >= 70:
        print("Grade: C | Good, Can Improve")
    elif score >= 60:
        print("Grade: D | Needs Improvement")
    else:
        print("Grade: F | Fail - Work Harder")
except ValueError:
    print("Error: Please enter a valid number.")
```

**Problem 2: Number Pattern Generator**

Write a function `print_pattern(n)` that takes a number `n` and prints this pattern:

For n = 4:

1

2 2

3 3 3

4 4 4 4

```python
def print_pattern(n):
    for i in range(1, n+1):
        for j in range(i):
            print(i, end=" ")
        print()  # New line

# Test
num = int(input("Enter number of rows: "))
print_pattern(num)
```

**Problem 3: Even-Odd Range Analyzer**

Write a program that takes two numbers (start and end) from the user and then:
- Prints all even numbers in the range
- Prints all odd numbers in the range
- Shows count of even and odd numbers

```python
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

even_count = 0
odd_count = 0

print("\nEven numbers:", end=" ")
for i in range(start, end + 1):
    if i % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print(f"\n\nTotal Even: {even_count}")
print(f"Total Odd: {odd_count}")
```
