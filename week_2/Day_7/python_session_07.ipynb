### Problem 1: Banking System

**Scenario**:  
Design a simple banking system with different types of accounts.

**Requirements**:
- Create a base class `BankAccount` with attributes: `account_number`, `holder_name`, `balance`
- Methods: `deposit()`, `withdraw()`, `display_balance()`
- Create two derived classes:
  - `SavingsAccount` (with `interest_rate` and `add_interest()` method)
  - `CurrentAccount` (with overdraft facility)
- Demonstrate polymorphism by handling multiple account types uniformly.

```python
class BankAccount:
    def __init__(self, account_number, holder_name, balance=0.0):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited. New balance: ₹{self.balance:.2f}")
        else:
            print("Invalid deposit amount!")
    
    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn. New balance: ₹{self.balance:.2f}")
        else:
            print("Insufficient balance or invalid amount!")
    
    def display_balance(self):
        print(f"[{self.account_number}] {self.holder_name}: ₹{self.balance:.2f}")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance=0.0, interest_rate=0.04):
        super().__init__(account_number, holder_name, balance)
        self.interest_rate = interest_rate
    
    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest of ₹{interest:.2f} added.")

class CurrentAccount(BankAccount):
    def __init__(self, account_number, holder_name, balance=0.0, overdraft_limit=5000):
        super().__init__(account_number, holder_name, balance)
        self.overdraft_limit = overdraft_limit
    
    def withdraw(self, amount):
        if amount > 0 and amount <= (self.balance + self.overdraft_limit):
            self.balance -= amount
            print(f"₹{amount} withdrawn. New balance: ₹{self.balance:.2f}")
        else:
            print("Exceeds overdraft limit!")

# Test
accounts = [SavingsAccount("S001", "Alice", 10000), CurrentAccount("C001", "Bob", 2000)]
for acc in accounts:
    acc.display_balance()
    acc.deposit(1000)
    acc.withdraw(3000)
    if isinstance(acc, SavingsAccount):
        acc.add_interest()
```

### Problem 2: Shape Calculator

**Scenario**:  
Create a geometry app to calculate area and perimeter of different shapes.

**Requirements**:
- Base class `Shape` with abstract methods `calculate_area()` and `calculate_perimeter()`
- Derived classes: `Rectangle`, `Circle`, `Triangle`
- Write a polymorphic function that accepts any shape and displays its info.

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    
    @abstractmethod
    def calculate_perimeter(self):
        pass
    
    def display_info(self):
        print(f"{type(self).__name__}:")
        print(f"   Area      : {self.calculate_area():.2f}")
        print(f"   Perimeter : {self.calculate_perimeter():.2f}\n")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def calculate_area(self):
        return self.length * self.width
    
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def calculate_area(self):
        return math.pi * self.radius ** 2
    
    def calculate_perimeter(self):
        return 2 * math.pi * self.radius

class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def calculate_area(self):
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
    
    def calculate_perimeter(self):
        return self.a + self.b + self.c

# Polymorphism Demo
shapes = [Rectangle(10, 6), Circle(5), Triangle(3, 4, 5)]
for shape in shapes:
    shape.display_info()
```

### Problem 3: Employee Management System

**Scenario**:  
Develop an HR system for different types of employees.

**Requirements**:
- Base class `Employee` with `calculate_salary()` and `get_details()`
- Derived classes: `Manager`, `Developer`, `Intern`
- Use method overriding and polymorphism to display salary information uniformly.

```python
class Employee:
    def __init__(self, emp_id, name, base_salary):
        self.emp_id = emp_id
        self.name = name
        self.base_salary = base_salary
    
    def calculate_salary(self):
        return self.base_salary
    
    def get_details(self):
        return f"{self.emp_id} | {self.name} | Salary: ₹{self.calculate_salary():.2f}"

class Manager(Employee):
    def __init__(self, emp_id, name, base_salary, bonus_percent=0.20):
        super().__init__(emp_id, name, base_salary)
        self.bonus_percent = bonus_percent
    
    def calculate_salary(self):
        return self.base_salary * (1 + self.bonus_percent)

class Developer(Employee):
    def __init__(self, emp_id, name, base_salary, projects=0):
        super().__init__(emp_id, name, base_salary)
        self.projects = projects
    
    def calculate_salary(self):
        return self.base_salary + (self.projects * 3000)

class Intern(Employee):
    def calculate_salary(self):
        return 10000  # Fixed stipend

# Polymorphic Function
def generate_payroll(employees):
    print("=== Company Payroll ===\n")
    for emp in employees:
        print(emp.get_details())

# Test
staff = [
    Manager("M001", "Priya Sharma", 120000),
    Developer("D042", "Rahul Verma", 85000, 4),
    Intern("I015", "Sneha Gupta", 0)
]

generate_payroll(staff)
```
