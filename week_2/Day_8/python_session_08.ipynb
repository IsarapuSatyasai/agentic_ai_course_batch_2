## Try & Except: Error Handling

**Definition**: Exception handling allows your code to gracefully handle runtime errors instead of crashing.

**Key Concepts**:
- `try`: Code that might raise an exception
- `except`: Catch specific or general exceptions
- `else`: Code that runs if no exception
- `finally`: Always runs (cleanup)
- `raise`: Re-raise or custom exceptions

**Best Practices**:
- Catch specific exceptions
- Don't silence errors silently
- Use context managers (`with`) for resources

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Cannot divide by zero!")
        return None
    except Exception as e:
        print(f"Unexpected error: {e}")
        return None
    else:
        print(f"Result: {result}")
        return result
    finally:
        print("Operation completed.")

safe_divide(10, 2)
safe_divide(10, 0)

# Custom exception
class ValidationError(Exception):
    pass

def validate_age(age):
    if age < 0:
        raise ValidationError(f"Invalid age: {age}")
    return age

try:
    validate_age(-10)
except ValidationError as e:
    print("Custom error:", e)
```

## Reading & Writing Files

**Best Practice**: Always use context managers (`with` statement) to ensure files are properly closed.

**Modes**:
- `'r'`: Read (default)
- `'w'`: Write (overwrites)
- `'a'`: Append
- `'rb'/'wb'`: Binary

```python
# Writing to a file
with open("session_notes.txt", "w") as f:
    f.write("Hello Python Students!\n")
    f.write("Welcome to Session 08.\n")
    f.writelines(["Learning file I/O today.\n"])

print("File written successfully!")

# Reading the file
with open("session_notes.txt", "r") as f:
    content = f.read()
    print(content)
```

## JSON Handling

JSON is the standard for data interchange.

```python
import json

data = {
    "name": "Alice",
    "age": 30,
    "skills": ["Python", "Data Science"]
}

# Writing JSON
with open("data.json", "w") as f:
    json.dump(data, f, indent=2)

print("JSON written to data.json")

# Reading JSON
with open("data.json", "r") as f:
    loaded = json.load(f)
    print(loaded)
```

## CSV Handling

Great for tabular data.

```python
import csv

# Writing CSV
with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Age", "City"])
    writer.writerow(["Alice", 30, "New York"])
    writer.writerow(["Bob", 25, "San Francisco"])
    writer.writerow(["Charlie", 35, "London"])

# Reading CSV
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(",".join(row))
```

## Dataclasses

**Definition**: Dataclasses are a decorator-based way to create classes that are primarily used to store data. Introduced in Python 3.7.

**Key Features**:
- Automatically generates `__init__`, `__repr__`, `__eq__`, and more
- Reduces boilerplate code
- Supports default values, type hints, and field customization

**Advantages**:
- Clean and readable
- Less error-prone than manual class writing
- Great for data containers (DTOs, records)

**Use Cases**:
- Configuration objects
- API response models
- Database entities

```python
from dataclasses import dataclass, field
from typing import List

@dataclass
class Person:
    name: str
    age: int
    email: str = ""
    hobbies: List[str] = field(default_factory=list)
    
    def introduce(self):
        return f"{self.name} is {self.age} years old"

# Creating instances
p1 = Person("Alice", 30, "alice@example.com")
p2 = Person("Bob", 25)

print(p1)
print(p2)
print(p1.introduce())
print(p1 == Person("Alice", 30, "alice@example.com"))  # Automatic __eq__
```

## Pydantic (Data Validation & Settings Management)

**Definition**: Pydantic is a Python library for data parsing, validation, and serialization using Python type annotations.

**Note**: Install with `!pip install pydantic` if not available.

**Key Features**:
- Automatic validation
- Rich error messages
- Serialization to/from JSON, dicts
- Support for nested models

**Advantages over plain dataclasses**:
- Runtime validation
- Parsing from untrusted sources (APIs, user input)
- Excellent for FastAPI, data pipelines

```python
try:
    from pydantic import BaseModel, Field, ValidationError
except ImportError:
    print("Pydantic not installed. Install with: pip install pydantic")
    # For demo, we'll skip or use placeholder
else:
    class User(BaseModel):
        name: str
        age: int = Field(ge=0)  # Greater than or equal to 0
        email: str
        is_active: bool = True
    
    # Valid data
    user = User(name="Alice", age=30, email="alice@example.com")
    print(user)
    
    # Invalid data example
    try:
        User(name="Bob", age=-5, email="bob@example.com")
    except ValidationError as e:
        print("Validation Error:", e)
```

## Summary: When to Use What?

| Concept              | Best For                              | Key Benefit                     |
|----------------------|---------------------------------------|---------------------------------|
| **Dataclass**        | Simple data containers                | Reduced boilerplate             |
| **Pydantic**         | Data validation & APIs                | Robust parsing & validation     |
| **Try/Except**       | Error handling                        | Graceful failure & recovery     |
| **File I/O**         | Persistent storage                    | Data persistence                |
| **JSON**             | Web APIs, config                      | Universal interoperability      |
| **CSV**              | Tabular data                          | Spreadsheet compatibility       |
