## Strings

**Definition**: Strings are immutable ordered sequences of Unicode characters used to represent text data.

**Key Characteristics**:
- Immutable (creating a "modified" string produces a new object)
- Ordered, iterable, and support indexing/slicing
- Rich set of built-in methods
- Multiple formatting options (f-strings recommended)

**Common Operations**:
- Concatenation, repetition, slicing
- `upper()`, `lower()`, `strip()`, `split()`, `join()`
- `find()`, `replace()`, `startswith()`, `endswith()`
- Formatting with f-strings

**Advantages**:
- Simple and readable
- Highly optimized
- Excellent for text processing

**Disadvantages**:
- Immutability can cause performance overhead in heavy concatenation loops (use `join()` instead)

**Use Cases**:
- User input/output, file I/O, logging
- Data cleaning and parsing
- Templating and report generation

```python
# Detailed String Examples
name = "ada lovelace"
greeting = "Hello"

print("Original:", name)
print("Upper:", name.upper())
print("Title:", name.title())
print("Capitalize:", name.capitalize())

# Stripping and cleaning
messy = "   Python is   awesome!   "
print("Stripped:", repr(messy.strip()))
print("Split:", messy.strip().split())

# Joining
words = ["Python", "is", "great"]
print("Joined:", " ".join(words))

# Formatting
age = 28
print(f"{name.title()} is {age} years old.")           # f-string (recommended)
print("Hello, {}!".format(name.title()))               # .format()

# Checking
print("Starts with 'Py':", "Python".startswith("Py"))
print("'on' in 'Python':", "on" in "Python")

# Slicing
print("First 3 chars:", "Python"[:3])
print("Reversed:", "Python"[::-1])
```

## OOP Fundamentals: Classes & Objects

**Definition**: 
- **Class**: Blueprint for creating objects.
- **Object**: Instance of a class (real entity with state and behavior).

**Key Concepts**:
- `__init__` (constructor)
- Attributes (instance variables)
- Methods (functions inside class)
- `self` parameter
- Encapsulation (bundling data and methods)

**Advantages**:
- Organizes code into reusable, modular units
- Models real-world entities naturally
- Supports maintainability and scalability

**Use Cases**:
- Representing entities (User, Product, Agent, etc.)
- Building libraries and frameworks
- Managing complex application state

```python
class Person:
    """Simple class example"""
    
    # Class attribute (shared by all instances)
    species = "Homo sapiens"
    
    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age
    
    # Instance method
    def introduce(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old."
    
    def have_birthday(self):
        self.age += 1
        return f"Happy birthday {self.name}!"

# Creating objects (instances)
p1 = Person("Alice", 30)
p2 = Person("Bob", 25)

print(p1.introduce())
print(p2.introduce())
print("Species:", p1.species)

p1.have_birthday()
print(p1.introduce())
```

## Inheritance & Polymorphism

**Inheritance**:
- Mechanism to create a new class (child) from an existing class (parent)
- Promotes code reuse ("is-a" relationship)

**Polymorphism** ("many forms"):
- Ability to use a single interface for different underlying forms (data types)
- Achieved via method overriding and duck typing in Python

**Key Benefits**:
- DRY principle (Don't Repeat Yourself)
- Extensible hierarchies
- Flexible runtime behavior

```python
# Base class
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some generic sound"
    
    def move(self):
        return f"{self.name} is moving"

# Child classes (Inheritance)
class Dog(Animal):
    def speak(self):                    # Method overriding
        return "Woof! Woof!"
    
    def fetch(self):
        return f"{self.name} is fetching the ball"

class Cat(Animal):
    def speak(self):
        return "Meow!"
    
    def move(self):                     # Overriding move
        return f"{self.name} is prowling gracefully"

# Polymorphism in action
def make_animal_speak(animal):
    print(animal.speak())               # Same interface, different behavior

animals = [Dog("Buddy"), Cat("Whiskers"), Animal("Generic")]

for animal in animals:
    print(f"\n{animal.name}:")
    print(animal.move())
    make_animal_speak(animal)

# Check inheritance
print("\nIs Dog subclass of Animal?", issubclass(Dog, Animal))
print("Is Buddy instance of Animal?", isinstance(animals[0], Animal))
```
